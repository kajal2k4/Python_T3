```python
import pandas as pd
request_url= f"http://api.openweathermap.org/data/2.5/forecast?lat=44.34&lon=10.99&appid={api_key}"
response = requests.get(request_url)
coordinates = response.json()
print(json.dumps(coordinates,indent=5))
datetime = []
temp=[]
press= []
humi=[]
desc=[]
for i in coordinates["list"] :
    datetime.append(i["dt_txt"])
    temp.append(i["main"]["temp"])
    press.append(i["main"]["pressure"])
    humi.append(i["main"]["humidity"])
    desc.append(i["weather"][0]["description"])
    
df = pd.DataFrame({"date_time":datetime,"temp":temp,"pressure":press,"humidity":humi,"weather":desc})
print(df)
```

```python
pollution_url = f"http://api.openweathermap.org/data/2.5/air_pollution/history?lat={lat}&lon={lon}&start=1606223802&end=1606482999&appid={api_key}"
pollution_response = requests.get(pollution_url)
pollution_data = pollution_response.json()

pollutants = pollution_data['list'][0]['components']

pollution_df = pd.DataFrame([pollutants])
print("\nCurrent air pollution data:")
print(pollution_df)
```
